```python
import os
import cv2
from google.colab import files

def resize_images(input_folder, output_folder, target_size):
    # Create output folder if not exists
    if not os.path.exists(output_folder):
        os.makedirs(output_folder)

    # Get list of files in input folder
    files_list = os.listdir(input_folder)

    for file_name in files_list:
        # Read image
        image_path = os.path.join(input_folder, file_name)
        img = cv2.imread(image_path)

        # Resize image
        resized_img = cv2.resize(img, target_size)

        # Write resized image to output folder
        output_path = os.path.join(output_folder, file_name)
        cv2.imwrite(output_path, resized_img)

def main():
    # Upload files to Colab
    print("Upload images to resize:")
    uploaded = files.upload()

    # Specify input and output folders
    input_folder = "/content/input_images"
    output_folder = "/content/output_images"

    # Create input folder if not exists
    if not os.path.exists(input_folder):
        os.makedirs(input_folder)

    # Move uploaded files to input folder
    for name, data in uploaded.items():
        with open(os.path.join(input_folder, name), "wb") as f:
            f.write(data)

    # Get target size for resizing
    target_width = int(input("Enter target width: "))
    target_height = int(input("Enter target height: "))
    target_size = (target_width, target_height)

    # Resize images
    resize_images(input_folder, output_folder, target_size)

    print("Images resized successfully!")

if __name__ == "__main__":
    main()
```
